```python
from jetbot import Robot, Camera
import cv2
import traitlets
import ipywidgets as widgets
from IPython.display import display
from jetbot import bgr8_to_jpeg

# Setup
robot = Robot()
camera = Camera.instance(width=224, height=224)
face_cascade = cv2.CascadeClassifier('/workspace/jetbot/notebooks/collision_avoidance/haarcascade_frontalface_default.xml')

# Image widget and toggle button
image = widgets.Image(format='jpeg', width=224, height=224)
enable_movement = widgets.ToggleButton(value=False, description='Enable Movement')

display(enable_movement)
display(image)

# Face detection and robot control
def detect_faces(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    #print("Faces detected:", len(faces), " | Movement enabled:", enable_movement.value)

    if enable_movement.value:
        if len(faces) == 1:
            print("→ 1 face → Turning LEFT")
            robot.left(0.3)
        elif len(faces) >= 2:
            print("→ 2+ faces → Turning RIGHT")
            robot.right(0.3)
        else:
            print("→ No face → STOP")
            robot.stop()
    else:
        robot.stop()

    # Draw rectangles for visual feedback
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    return bgr8_to_jpeg(frame)

# Only ONE camera link, to ONE image widget
camera_link = traitlets.dlink((camera, 'value'), (image, 'value'), transform=detect_faces)
```
